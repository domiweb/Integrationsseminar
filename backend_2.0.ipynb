```python
import pandas as pd
import re
import numpy as np
```

### Data Cleaning

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
symptoms = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
```

```python
symptoms
```

```python
del symptoms['Symptom_7']
del symptoms['Symptom_8']
del symptoms['Symptom_9']
del symptoms['Symptom_10']
del symptoms['Symptom_11']
del symptoms['Symptom_12']
del symptoms['Symptom_13']
del symptoms['Symptom_14']
del symptoms['Symptom_15']
del symptoms['Symptom_16']
del symptoms['Symptom_17']
```

```python
symptoms = symptoms.fillna(' ')
```

```python
symptoms
```

```python
target_column = 'Disease'
feature_columns = ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4', 'Symptom_5', 'Symptom_6']
```

```python
def str_lower(s):
  return s.lower()

symptoms[target_column] = symptoms[target_column].apply(str_lower)
```

```python
def remove_punct(s):
  return s.replace('_', ' ')

for col in feature_columns:
  symptoms[col] = symptoms[col].apply(remove_punct)
```

```python
symptoms
```

### Preprocessing

```python
def str_to_vector(s):
  z = 3047.4551245522744
  n = 4920
  return len(s)*z/n

for column in feature_columns:
  symptoms[column] = symptoms[column].apply(str_to_vector)
```

```python
symptoms
```

```python
train_X = symptoms[feature_columns]
train_Y = symptoms[target_column]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
train_Y = le.fit_transform(train_Y)
```

```python
X = train_X.values
y = train_Y
```

### Model

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers.core import Dense, Activation
from keras.optimizers import Adam
from keras.models import Sequential
```

```python
healthBot = Sequential()
healthBot.add(Dense(10, input_dim=X.shape[1]))
healthBot.add(Activation('relu'))
healthBot.add(Dense(20))
healthBot.add(Activation('relu'))
healthBot.add(Dense(15))
healthBot.add(Activation('relu'))
healthBot.add(Dense(20))
healthBot.add(Activation('relu'))
healthBot.add(Dense(1))
healthBot.add(Activation('relu'))

healthBot.compile(metrics=['accuracy'], optimizer=Adam(learning_rate=0.1), loss='categorical_crossentropy')
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='loss', patience=3, min_delta=1e-6)
```

```python
hist = healthBot.fit(X, y, epochs=1000, batch_size=32, 
              #callbacks=early_stopping
              )
```

```python
test = str_to_vector('itching')
test1 = str_to_vector('skin rash')

l = [test, test1]
```
